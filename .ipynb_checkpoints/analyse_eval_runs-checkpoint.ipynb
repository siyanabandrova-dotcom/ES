The best run without considering AIME25 took ~20h, at the first eval step, and used:

The best run without considering AIME25 and using population size 4096 took ~18h53m, at the first eval step, and used:

- dataset: orz57k
- population size: 4,096
- normalize-with-std
- max tokens: 8,192

It achieves the same average score as the second best run (which is identical, except that it uses deepscaler40k to train), but it takes around 13k seconds (~3h36m) less to train.

Moreover, the best run with population size 4,096 is the 4th best run overall (when not using AIME25), achieving 0.025 less in the composite score.

## Population Size = 256

```python
best_eval_noaime25_p256 = pd.read_csv("rebuttals/global_top_runs_no_aime25_p256.csv")
best_eval_noaime25_p256
```

```python
best_eval_noaime25_p256.iloc[0]["run"]
```

The best run without considering AIME25 and using population size 256 took ~21h53m, at the second eval step, and used:

- dataset: orz57k
- population size: 256
- normalize-with-std
- max tokens: 4,096

The difference in composite score between smaller populations and larger populations is quite notable, with a difference of more than 0.1.

It must be noted that there is a certain bias: when training with a larger population sizes, it seems that the model learns to be more succint and more accurate. This means that as training progresses, training iterations last for less. With population size 256, the speed of learning falls, which means that none of the runs which used max tokens 8192 were able to finish.

- dataset: orz57k

- no normalize-with-std
- max tokens: 4,096

The difference in composite score between using normalize-with-std and not is extremely significant. Even at the largest population size, the composite score is smaller than a run using population size 256.
